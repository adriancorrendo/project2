```python
import numpy as np
```

```python
def vpdfun(T_grid, RH_grid):
    """Function to estimate the vapor pressure deficit (vpd) with air temperature (T, Celsius)
    and relative humidity (RH, %) as inputs.
    Outputs: vpd in KPa 
    Author: Adrian Correndo
    Date: March-04-2019"""
    e_sat = 0.611 *np.exp((17.5*T_grid/(T_grid+241)))
    e_act = e_sat * (RH_grid/100)
    vpd = e_sat - e_act
    return vpd
```

```python
T = np.linspace(0,40,5)#Temperature in Celsius
RH = np.linspace(0,100,5)#relative humidity in percentage
T_grid, RH_grid = np.meshgrid(T,RH) # Creating a grid of each element for matrix operations
#print(T_grid) # for checking square format of T_grid
#print(RH_grid) # for checking square format of RH_grid
```

```python
VPD = vpdfun(T_grid, RH_grid)
print(VPD)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
%matplotlib inline

#Create a Figure
fig = plt.figure(figsize=(10, 6), dpi=80) #10" by 6" dpi=dots per inch
#Get figure axes and convert it into a 3D projection
ax = fig.gca(projection='3d')

# Add surface plt to axes. Save this surface plot in a variable
surf = ax.plot_surface(RH_grid, T_grid, VPD, camp=cm.coolwarm, linewidth=0)

#add colorbar to figure based on ranges in the surf map.
#Colorbar belongs to the figure.
#The values of the colorbar are from the axes, this why we need to pass the ...
fig.colorbar(surf, shrink=0.5, aspect=5)

# Label x, y, z axis
ax.set_xlabel("Relative Humidity [%]")
ax.set_ylabel("Air Temperature [\N{DEGREE SIGN}C]")
ax.set_zlabel("VDP [KPa]")

# Set the position of the 3D plot
ax.view_init(elev=20, azim=20) #Elevation and azimuth. Change their values to...

plt.show()
```
