# Subplots

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dirname = '/Users/correndo/Desktop/Coding/introcoding-spring-2019/Datasets/'
filename = 'global_wheat.csv'
```

```python
df = pd.read_csv(dirname + filename)
df.head()
```

```python
# Subplots
plt.figure()
plt.subplot(2,1,1)
plt.plot(df.year, df.Argentina_yield_hg_ha)
plt.ylabel("Yield kg/ha")
plt.subplot(2,1,2)
plt.plot(df.year, df.US_yield_hg_ha)
plt.ylabel("Yield kg/ha")

plt.show()
```
